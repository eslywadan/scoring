```python
# Describe the behavior of essay_to_sentences()
essay_v = tsm.X['essay'][0]
print(f"Step 1: Original essay {essay_v}")
raw_sentences = tokenizer.tokenize(essay_v.strip()) 
print(f"Step 2: Raw sentence after tokenization {raw_sentences}")
## the tokenize function return a list containing sigle sentence as element
# 每一段句字是可以獨立存在、有意義的單一子句。 所以此動作應該視為斷句
_words = []
for raw_sentence in raw_sentences:
        if len(raw_sentence) > 0:
            _words.append(essay_to_wordlist(raw_sentence, remove_stopwords=True))
print(f"Step 3: {_words}")
## Step 3 為斷字，在此步驟，每一個句字，都會被斷成單字。
```

```python
# Train the english word2vec model
sentences = tsm.essays_to_sentences(tsm.X['essay'])
print(type(sentences))
print(sentences[0])

num_features = 300  # 特征向量的维度
min_word_count = 40  # 最小词频，小于min_word_count的词被丢弃
num_workers = 4  # 训练的并行数
context = 10 # 当前词与预测词在一个句子中的最大距离
downsampling = 1e-3 # 高频词汇的随机降采样的配置阈值

# 训练模型
print("Training Word2Vec Model...")
model = Word2Vec(sentences, workers=num_workers, vector_size=num_features, min_count=min_word_count, window=context,
                            sample=downsampling)
model.init_sims(replace=True)  # 结束训练后锁定模型，使模型的存储更加高效
model.save('word2vecmodel_eng.model') # 保存模型
```

```python
sentences_2 = tsm_tayal.essays_to_sentences(tsm_tayal.X['essay'])
print(type(sentences_2))
print(sentences_2)

num_features = 10  # 特征向量的维度
min_word_count = 2  # 最小词频，小于min_word_count的词被丢弃
num_workers = 1  # 训练的并行数
context = 10 # 当前词与预测词在一个句子中的最大距离
downsampling = 1e-3 # 高频词汇的随机降采样的配置阈值

# 训练模型
print("Training Word2Vec Model...")
model = Word2Vec(sentences_2, workers=num_workers, vector_size=num_features, min_count=min_word_count, window=context,
                            sample=downsampling)
model.init_sims(replace=True)  # 结束训练后锁定模型，使模型的存储更加高效
model.save('word2vec_tayal1.model')
# model.wv.save_word2vec_format('word2vecmodel_tayal1.bin', binary=True) # 保存模型
```
